# rolling特徴量の作成

```python
import pickle
import pandas as pd
import numpy as np
```

```python
with open('../data/pickle/diff_time_data.pkl', mode='rb') as f:
    df = pickle.load(f)
```

```python
df.head()
```

```python
df.index = df.click_time
```

```python
df.click_time = pd.to_datetime(df.click_time, unit='s')
df.index = pd.to_datetime(df.index, unit='s')
```

```python
df['diff'] = df['diff'] * -1
```

```python
grouped = df.groupby('ip').app.rolling('300s').count()
```

```python
grouped = grouped.groupby(['ip', 'click_time']).max()
```

```python
windowed_df = grouped.reset_index()
```

```python
merged = pd.merge(df, windowed_df, on=['ip', 'click_time'])
```

```python
merged.columns = ['app', 'attributed_time', 'channel', 'click_id', 'click_time', 'device', 'ip', 'is_attributed', 'os', 'diff', 'window_count']
```

```python
len(merged)
```

```python
pickle.dump(merged[['ip', 'click_time','window_count']], open("../data/pickle/add_rolling_and_diff_feature.pkl", 'wb'), protocol=4)
```

```python
max(merged.window_count)
```
