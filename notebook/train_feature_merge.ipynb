```python
import pandas as pd
import pickle
```

```python
dtypes = {
    'ip': 'uint32',
    'app': 'uint16',
    'device': 'uint16',
    'os': 'uint16',
    'channel': 'uint16',
    'is_attributed': 'uint8',
    'click_id': 'uint32'
}
```

```python
train_supplement_df = pd.read_hdf('../tamaki_script/X_train_add_supplement.h5', 'table')
with open('../data/pickle/add_rolling_and_diff_feature.pkl', mode='rb') as f:
    rolling_df = pickle.load(f)
train_supplement_df.click_time = pd.to_datetime(train_supplement_df.click_time, unit='s')
```

```python
train_supplement_df = pd.merge(train_supplement_df, rolling_df[['ip', 'click_time', 'window_count']], on=[
    'ip', 'click_time'], how='left')
```

```python
train_supplement_df = train_supplement_df.drop_duplicates()
```
